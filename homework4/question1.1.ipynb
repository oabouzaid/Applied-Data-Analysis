# ` 1 Propensity Score Matching`

## `1(1) A Naive Analysis`

### `Load the Data`

```python
data = pd.read_csv("data/lalonde.csv")
```

### `Compare the distribution of the outcome variable` *`re78`* `between the two groups`

```python
re78_group1 = data[data.treat==1]["re78"]
re78_group2 = data[data.treat==0]["re78"]
```

### `Using Plots`

```python
sns.distplot(re78_group1)
```

```python
sns.distplot(re78_group2)
```

```python
sns.distplot(re78_group1)
sns.distplot(re78_group2)
```

`Both of the graphs seem like Poisson Distribution with` $\lambda \approx 1$

### `Using Numbers`

```python
re78_group1.describe()
```

```python
re78_group2.describe()
```

**Kolmogorov–Smirnov Test**

```python
stats.ks_2samp(re78_group1, re78_group2)
```

$p-value \geqslant 0.05$ `so, we reject the null hypothesis meanig that there is no difference betwen distribtuion of the variable between two groups.`
